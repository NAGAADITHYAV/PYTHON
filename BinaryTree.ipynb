```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.parent = None
        self.left = None
        self.right = None
```

```python
class BinaryTree:
    def __init__(self, root = None):
        self.root = root
    
    def root(self):
        return self.root
    
    def height(self):
        '''Return the height of tree'''
        return self._height(self.root)
    
    def _height(self, node):
        if not node:
            return 0
        else:
            return (1 + max(self._height(node.left), self._height(node.right)))

    def subtree_first(self, node):
        '''subtree with given node as a root and its left most leaf'''
        if not node:
            return node
        while node.left:
            node = node.left
        return node
    
    def subtree_last(self, node):
        '''subtree with given node as a root and its left most leaf'''
        if not node:
            return node
        while node.right:
            node = node.right
        return node
    
    def successor(self, node):
        '''finding sucessor of a node'''
        if node.right:
            return self.subtree_first(node.right)
        while node.parent:
            if node.parent.left == node:
                return node.parent
            else:
                node = node.parent
        return None
    
    def predecessor(self, node):
        '''finding predecessor of a node'''
        if node.left:
            return self.subtree_last(node.left)
        
        return node.parent
    
    def subtree_insert_after(self, node, new_node):
        '''insert after a given node in traversal order'''
        if not node.right:
            node.right = new_node
            new_node.parent = node
        else:
            node = self.subtree_first(node)
            node.left = new_node
            new_node.parent = node
    
    def subtree_insert_before(self, node, new_node):
        '''insert before a given node in traversal order'''
        if not node.left:
            node.left = new_node
            new_node.parent = node
        else:
            node = self.subtree_last(node)
            node.right = new_node
            new_node.parent = node
    
    def delete(self, node):
        node = self._swap_recursive_with_beside(node)
        self._delete_leaf(node)

    def _swap_recursive_with_beside(self, node):
        if not node.right and not node.left:
            return node
        else:
            if node.left:
                side_node = self.predecessor(node)
            else:
                side_node = self.successor(node)
            side_node.data, node.data = node.data, side_node.data
            return self._swap_recursive_with_beside(side_node)

    def _delete_leaf(self, node):
        if not node.parent:
            self.root = None
        else:
            new_leaf = node.parent
            if new_leaf.left == node:
                new_leaf.left = None
            else:
                new_leaf.right = None
    
```
