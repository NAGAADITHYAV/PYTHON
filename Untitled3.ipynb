```python
def maxEqualFreqSubarray(arr):
    freq = {}
    countFreq = {}
    maxLen = 0
    start = 0

    for end in range(len(arr)):
        num = arr[end]
        
        # Update the frequency map
        if num in freq:
            oldFreq = freq[num]
            freq[num] += 1
        else:
            oldFreq = 0
            freq[num] = 1
        
        # Update countFreq map
        if oldFreq > 0:
            countFreq[oldFreq] -= 1
            if countFreq[oldFreq] == 0:
                del countFreq[oldFreq]
        
        newFreq = freq[num]
        if newFreq in countFreq:
            countFreq[newFreq] += 1
        else:
            countFreq[newFreq] = 1
        
        # Check if current window is valid
        maxFreq = max(countFreq.keys())
        minFreq = min(countFreq.keys())

        if len(countFreq) == 1:
            # All elements have the same frequency
            if maxFreq == 1 or countFreq[maxFreq] == len(arr[start:end+1]):
                maxLen = max(maxLen, end - start + 1)
        elif len(countFreq) == 2:
            # Check two cases:
            # 1. One frequency has count 1 and can be removed
            # 2. One element occurs once more than the rest and has a count of 1
            if (countFreq[maxFreq] == 1 and maxFreq - minFreq == 1) or (countFreq[minFreq] == 1 and minFreq == 1):
                maxLen = max(maxLen, end - start + 1)

        # Move the window if it's invalid
        while len(countFreq) > 2:
            numStart = arr[start]
            startFreq = freq[numStart]
            freq[numStart] -= 1

            countFreq[startFreq] -= 1
            if countFreq[startFreq] == 0:
                del countFreq[startFreq]
            
            if freq[numStart] > 0:
                if freq[numStart] in countFreq:
                    countFreq[freq[numStart]] += 1
                else:
                    countFreq[freq[numStart]] = 1
            else:
                del freq[numStart]
            
            start += 1
    
    return maxLen

# Example usage:
arr = [1, 2, 1, 3, 4, 2, 4, 3, 3, 4]
print(maxEqualFreqSubarray(arr))  # Output: 8
```

```python
def max_subarr_size(n, a):
    res = 0
    freq_map = {}
    freq_count = {}
    l = 0

    for r in range(n):
        prev_count = freq_map.get(a[r], 0)
        freq_map[a[r]] = prev_count + 1

        if prev_count in freq_count:
            freq_count[prev_count] -= 1
            if freq_count[prev_count] == 0:
                del freq_count[prev_count]

        freq_count[prev_count + 1] = freq_count.get(prev_count + 1, 0) + 1

        while len(freq_count) > 1:
            left_elem = a[l]
            left_count = freq_map[left_elem]
            freq_count[left_count] -= 1
            if freq_count[left_count] == 0:
                del freq_count[left_count]
            freq_map[left_elem] -= 1
            l += 1

        if len(freq_count) == 1:
            res = max(res, r - l + 1)

    return res

# Driver Code
arr = [9,9,9]
n = len(arr)
print(max_subarr_size(n, arr))
```

```python
from collections import defaultdict

def find_special_nodes(n, edges):
    adj_list = defaultdict(set)
    for u, v in edges:
        adj_list[u].add(v)
        adj_list[v].add(u)
    
    special_nodes = [-1] * n
    for i in range(n):
        neighbors_i = adj_list[i]
        max_intersection_size = -1
        special_node = -1
        
        for j in range(n):
            if j == i or j in neighbors_i:
                continue
            
            neighbors_j = adj_list[j]
            intersection_size = len(neighbors_i & neighbors_j)
            
            if intersection_size > max_intersection_size or (intersection_size == max_intersection_size and j < special_node):
                max_intersection_size = intersection_size
                special_node = j
        
        special_nodes[i] = special_node
    
    return special_nodes

# Example usage:
n = 3  # Number of nodes
edges = [[0, 1], [0, 2], [2, 1]]  # List of edges in the graph
# n = 5
# fr = [[0, 1], [0, 2], [1, 3], [2, 3], [3, 4]]

special_nodes = find_special_nodes(n, edges)
print(special_nodes)
```
