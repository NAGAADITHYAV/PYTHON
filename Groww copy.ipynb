```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# res
```

```python
x = []
y1 = []
y2 = []
y3 = []
for ep in res:
    x.append(ep['epoch']) # x << ep['epoch']
    y1.append(ep['netWorth'])
    y2.append(ep['investedAmount'])
    y3.append(ep['netWorth']-ep['investedAmount'])

plt.plot(x, y1, label ='current corpus')
plt.plot(x, y2, '-.', label ='invested')
plt.plot(x, y3, '.', label = 'profit')

plt.xlabel("Epochs")
plt.ylabel("corpus")
plt.legend()
plt.title('multiple plots')
plt.show()
```
