```python
def compound_int(p,r,t):
    return p*(((100+r)/100)**t)

def simple_int(p,r,t):
    return p+((p*t*r)/100)

def simple_int_coumpounded(p,r,t,n):
    tp = t//n
    tl = t%n
    p = p*(((100+(n*r))/100)**tp)
    
    return p+((p*tl*r)/100)
```

```python
x,c,s,sc = [],[],[],[]

for i in range(0,500):
    x.append(i)
    c.append(compound_int(1,18,i))
    s.append(simple_int(1,18,i))
    sc.append(simple_int_coumpounded(1,18,i,3))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.plot(x, c, label ='c')
plt.plot(x, s, '-.', label ='s')
plt.plot(x, sc, '+', label ='sc')

plt.xlabel("years")
plt.ylabel("amount to be paid")
plt.legend()
plt.title('multiple plots')
plt.show()
```
