```python
response = requests.get("https://groww.in/v1/api/aggregator/v1/insight/net_worth/fetch_graph_data/ALL", headers = Headers)
```

```python
res = response.json()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
res
```

```python
x = []
y1 = []
y2 = []
y3 = []
for ep in res:
    x.append(ep['epoch']) # x << ep['epoch']
    y1.append(ep['netWorth'])
    y2.append(ep['investedAmount'])
    y3.append(ep['netWorth']-ep['investedAmount'])

plt.plot(x, y1, label ='current corpus')
plt.plot(x, y2, '-.', label ='invested')
plt.plot(x, y3, '_', label = 'profit')

plt.xlabel("Epochs")
plt.ylabel("corpus")
plt.legend()
plt.title('multiple plots')
plt.show()
```

```python
x = []
y = []
for ep in res:
    x.append(ep['epoch'])
    y.append((ep['netWorth']-ep['investedAmount'])/ep['investedAmount'])

plt.plot(x, y, label ='y')
plt.grid(which='both')

plt.xlabel("Epochs")
plt.ylabel("profs")
# plt.legend()
plt.title('profit plots')
plt.show()
```
