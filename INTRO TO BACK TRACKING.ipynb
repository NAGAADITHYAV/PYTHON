```python
class Solution:
    def generateParenthesis(self, n: int):
        stack = []
        res = []
        def backtracking(openn, closedn):
            print("###################################")
            print("level ", openn, closedn)
            print('stack ', stack)
            print('result until now', res)
            
            if openn == closedn == n:
                res.append(''.join(stack))
                print('appended result', ''.join(stack))
                return
            
            if openn < n:
                stack.append('(')
                print('STACK APPENDED', stack, 'at level ', openn, closedn)
                backtracking(openn+1, closedn)
                stack.pop()
                print('stack poped', stack, 'at level ', openn, closedn)
            
            if closedn < openn:
                stack.append(')')
                print('STACK APPENDED', stack, 'at level ', openn, closedn)
                backtracking(openn, closedn+1)
                stack.pop()
                print('stack poped', stack, 'at level ', openn, closedn)
        backtracking(0,0)
        return res

Solution().generateParenthesis(3)
```
